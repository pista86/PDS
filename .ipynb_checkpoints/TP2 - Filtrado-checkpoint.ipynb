### Archivo *ECG_TP4.mat*
(variables)
 - **ecg_lead**: Registro de ECG muestreado a $fs=1$ KHz durante una prueba de esfuerzo
 - **qrs_pattern1**: Complejo de ondas QRS normal
 - **heartbeat_pattern1**: Latido normal
 - **heartbeat_pattern2**: Latido de origen ventricular
 - **qrs_detections**: vector con las localizaciones (en # de muestras) donde ocurren los latidos
 
a) Establezca una plantilla de diseño para los filtros digitales que necesitará para que la señal de ECG se asemeje a los latidos promedio en cuanto a *suavidad* de los trazos y nivel isoeléctrico nulo.
 

```python
import numpy as np
# If the code is in a file called plot_zplane.py
from plot_zplane import zplane 
from scipy import signal
import matplotlib.pyplot as plt
import scipy.io as scio
import scipy.fftpack as sc

mat = scio.loadmat('ECG_TP4.mat')

ecg_lead = mat['ecg_lead']
qrs_pattern1 = mat['qrs_pattern1']
heartbeat_pattern1 = mat['heartbeat_pattern1']
heartbeat_pattern2 = mat['heartbeat_pattern2']
qrs_detections = mat['qrs_detections']

fs=1000


plt.figure
plt.plot(ecg_lead[200:10200])
plt.title('ecg_lead', fontsize=16)
plt.show()

plt.figure
plt.plot(qrs_pattern1)
plt.title('qrs_pattern1', fontsize=16)
plt.show()

plt.figure
plt.plot(heartbeat_pattern1)
plt.title('heartbeat_pattern1', fontsize=16)
plt.show()

plt.figure
plt.plot(heartbeat_pattern2)
plt.title('heartbeat_pattern2', fontsize=16)
plt.show()



# Análisis de espectro de latidos normales, señal de corta duración, se agregan ceros

zQRS = np.zeros(len(qrs_pattern1)*10)
zheartbeat_pattern1 = np.zeros(len(heartbeat_pattern1)*10)
zheartbeat_pattern2 = np.zeros(len(heartbeat_pattern2)*10)

qrs_pattern1_z = np.append(qrs_pattern1,zQRS)
heartbeat_pattern1_Z = np.append(heartbeat_pattern1,zheartbeat_pattern1)
heartbeat_pattern2_Z = np.append(heartbeat_pattern2,zheartbeat_pattern2)


N = len (qrs_pattern1_z)
spectrum = (2/N)*np.abs(sc.fft(qrs_pattern1_z))
half = spectrum[:N//2]
frec = np.linspace(0.0, fs//2, N//2)

plt.stem(frec, half)
plt.xlabel('Frecuencia [Hz] ', fontsize=14)
plt.ylabel('Amplitud normalizada', fontsize=14)
plt.title('Espectro qrs_pattern1_z', fontsize=16)
plt.show() 


print('Energía qrs_pattern1_z:', np.sum(half*half))

print('90% Energía qrs_pattern1_z tomando hasta f=10Hz:',np.sum(half[0:29]*half[0:29]))
print('f=',((fs/2)/len(half))*29 ,'Hz')


N = len (heartbeat_pattern1_Z)
spectrum = (2/N)*np.abs(sc.fft(heartbeat_pattern1_Z))
half = spectrum[:N//2]
frec = np.linspace(0.0, fs//2, N//2)

plt.stem(frec, half)
plt.xlabel('Frecuencia [Hz] ', fontsize=14)
plt.ylabel('Amplitud normalizada', fontsize=14)
plt.title('Espectro heartbeat_pattern1_Z', fontsize=16)
plt.show() 

print('Energía heartbeat_pattern1_Z:', np.sum(half*half))

print('90% Energía heartbeat_pattern1_Z tomando hasta f=10Hz:',np.sum(half[0:90]*half[0:90]))
print('f=',((fs/2)/len(half))*90 ,'Hz')

N = len (heartbeat_pattern2_Z)
spectrum = (2/N)*np.abs(sc.fft(heartbeat_pattern2_Z))
half = spectrum[:N//2]
frec = np.linspace(0.0, fs//2, N//2)

plt.stem(frec, half)
plt.xlabel('Frecuencia [Hz] ', fontsize=14)
plt.ylabel('Amplitud normalizada', fontsize=14)
plt.title('Espectro heartbeat_pattern2_Z', fontsize=16)
plt.show() 


print('Energía heartbeat_pattern2_Z:', np.sum(half*half))

print('90% Energía heartbeat_pattern2_Z tomando hasta f=8Hz:',np.sum(half[0:55]*half[0:55]))
print('f=',((fs/2)/len(half))*55 ,'Hz')


```
