## Muestreo de señales - Fourier


<div class="alert alert-block alert-info">
<b>1)</b> Se busca realizar una librería en Python que nos permita obtener $N$ muestras de las siguientes señales de período $T_0=1/f_0$, muestreadas a una frecuencia $f_s$ :

</div>


+ **Senoidal**

> $s(k)=A_0.sin(2\pi.k.\frac{ f_0}{f_s} + \phi_0),\space\space \forall k \in \Bbb{R} \wedge k \in (0, N-1)$

+ __Ruido__ incorrelado de distribución Gaussiana y parámetros $\mu$ y $\sigma^2$

***

+ (Opcional) __Cuadrada__ con ciclo de actividad $d \in (0,1)$

> $s(k)=A_0,\space\space \forall k \in (0, N.d ]$ 

> $s(k)=-A_0,\space \forall k \in (N.d, N-1)$

+ (Opcional) __Triangular__ con punto de simetría $d \in (0,1)$ y coeficientes $a_{ij} \in \Bbb{R}$

> $s(k)=a_{11}.k/f_s+a_{12},\space\space \forall k \in (0, N.d ]$ 

> $s(k)=a_{21}.k/f_s+a_{22},\space \forall k \in (N.d, N-1)$


Es decir que se podría invocar la señal que genere la senoidal de la siguiente manera:


```python

def generador_senoidal (fs, f0, N, a0=1, p0=0):
    """ 
    
    brief:  Generador de señales senoidal, con argumentos
    
    fs:     frecuencia de muestreo de la señal [Hz]
    N:      cantidad de muestras de la señal a generar
    f0:     frecuencia de la senoidal [Hz]
    a0:     amplitud pico de la señal [V]
    p0:     fase de la señal sinusoidal [rad]
    
    como resultado la señal devuelve:
    
    signal: senoidal evaluada en cada instante 
    tt:     base de tiempo de la señal
    """    

    # comienzo de la función

    tt = np.linspace(0, (N-1)/fs, N)
        
    signal = a0 * np.sin(2*np.pi*f0*tt + p0)
    
    # fin de la función
    
    return tt, signal
```

<b>2.a)</b> (Opcional) Implemente un algoritmo que calcule la transformada discreta de Fourier (DFT). Utilice las funciones generadoras de señales de 1) para evaluar y ejemplificar su funcionamiento.
</div>

```python
def DFT(x): 
    N = np.size(x)
    X = np.zeros((N,),dtype=np.complex128)
    for m in range(0,N):    
        for n in range(0,N): 
            X[m] += x[n]*np.exp(-np.pi*2j*m*n/N)
    return X


N  = 1000 # muestras
fs = 1000 # Hz
a0 = 1     # Volts
p0 = 0     # radianes
f0 = 200 # Hz


[tt,signal] = generador_senoidal(fs, f0, N, a0, p0)

plt.figure()
plt.plot(tt,signal)
plt.xlabel('tiempo [segundos]', fontsize=14)
plt.ylabel('Amplitud [V]', fontsize=14)
plt.title('Señal: senoidal', fontsize=16)


dft = (2/N)*abs(DFT(signal))
frec = np.linspace(0.0,fs/2,N//2)
half_senal_dft = dft[:N//2]


plt.figure()
plt.plot(frec,half_senal_dft)
plt.xlabel('frecuencia [Hz]', fontsize=14)
plt.ylabel('Amplitud normalizada', fontsize=14)
plt.title('Señal: senoidal', fontsize=16)



```

```python
import time
import scipy.fftpack as sc

fs = 1000 # Hz
a0 = 1     # Volts
p0 = 0     # radianes
f0 = 200 # Hz

N_list= [16, 32, 64, 128, 256, 512, 1024, 2048]
T_dft_list = list()
T_fft_list = list()


for i in N_list:
    
    [tt,signal] = generador_senoidal(fs, f0, i, a0, p0)
    
    tinicial = time.time()    
    DFT(signal)
    T_dft_list.append(time.time() - tinicial)
    
    tinicial = time.time()    
    sc.fft(signal)
    T_fft_list.append(time.time() - tinicial)
    

plt.plot(N_list, T_dft_list,label='DFT')
plt.plot(N_list, T_fft_list,label='FFT')
plt.xlabel('Muestras ', fontsize=14)
plt.legend(fontsize=12, ncol=1)
plt.ylabel('Tiempo de procesamiento [S]', fontsize=14)
plt.title('Tiempos de procesamiento de DFT y FFT', fontsize=16)

```

**3.e)** Analice cuál es la frecuencia  $\hat{f}_0 = \mathop{arg\ max}_f \{\lvert X(f) \rvert \}$ a la que ocurre el máximo del espectro y en cuánto difiere de $f_0$ en términos porcentuales.

```python
N=1024
N_array = np.array([N//10,N,10*N])
fs = 1024 # Hz
a0 = 1     # Volts
p0 = 0     # radianes
f0 = fs/4 + 0.5

center_of_mass = list()
modulo_Fcentral_list = list()
diff_percent_list = list()

for i in N_array:

    [tt,signal] = generador_senoidal(fs, f0, i, a0, p0)

    np.pad(signal, (0, i), 'constant')

    spectrum = (2/i)*np.abs(sc.fft(signal))
    half = spectrum[:i//2]
    center_of_mass.append( ((half*np.arange(len(half))).sum()/half.sum()) * N/i )
    modulo_Fcentral_list.append( max(half) )
    Max_pos = np.where(half == modulo_Fcentral_list[-1]) 
    Max_pos = Max_pos[0] * N/i
    diff_percent_list.append( 100*(abs(center_of_mass[-1] - Max_pos[-1])/center_of_mass[-1]) )  
```
